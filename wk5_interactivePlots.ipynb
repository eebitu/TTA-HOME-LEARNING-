```python
import altair as alt
import pandas as pd
# Import seaborn
import seaborn as sns
```

```python
# Check out available datasets
print(sns.get_dataset_names())
```

```python
# Load my data
data = sns.load_dataset('tips')
```

```python
data
```

```python
brush = alt.selection_interval()
```

```python
points = alt.Chart(data).mark_point().encode(
    x='total_bill:Q',
    y='tip:Q',
    opacity=alt.value(0.2),
    color=alt.value('red')

)
points
```

```python
bars = alt.Chart(data).mark_bar().encode(
    y='size:N',
    color='size:N',
    x='count(size):Q'
)

bars
```

```python
## The largest size in this dataset is the size 2 of  the customers.
```

```python
#interarctive chart of the data
int_tips = alt.Chart(data).mark_circle(filled=True).encode(
         alt.X('total_bill'),
         alt.Y('tip'),
         alt.Size('size'),
         alt.Color('sex'),
         alt.OpacityValue(0.7),
         tooltip = [alt.Tooltip('day'),
         alt.Tooltip('total_bill'),
         alt.Tooltip('tip'),
         alt.Tooltip('smoker'),
         alt.Tooltip('time'),
         alt.Tooltip('size'),
         alt.Tooltip('sex')]   
        ).interactive()

int_tips
```

```python
## The interactive chart above displays all the features in the tip dataset. The x axis displays the total bill and the y axis displays the tip from the customers. 
## The highest tip and bill comes from a male customer with the highest bill of Â£50 who gives a tip of of Â£10. This male customer ate dinner on Saturday, smokes and is in the size 3 category.  
## The lowest tip and bill mix if from a female customer with a total bill Â£3.07 and gives a tip of Â£1. This customer ate dinner on a saturday, is a smoker and in the size 1 category.
```
